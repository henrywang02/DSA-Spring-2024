# Tree demo 1 - Binary tree traversal

*Description*: Let's create a binary tree and add five nodes into it (root node and four child nodes). Tree should then be traversed using pre-, in- and post-order while printing the content for the user.

```python
# Class for tree node
class Node:
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None

# Class for tree
class Tree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def insert(self, data):
        new_node = Node(data)
        if self.root is None:
            self.root = new_node
        else:
            self.insertNode(self.root, new_node)

    def insertNode(self, node, new_node):
        if new_node.data < node.data:
            if node.left is None:
                node.left = new_node
            else:
                self.insertNode(node.left, new_node)
        else:
            if node.right is None:
                node.right = new_node
            else:
                self.insertNode(node.right, new_node)

    def inorder(self, node):
        if node is not None:
            self.inorder(node.left)
            print(node.data)
            self.inorder(node.right)

    def preorder(self, node):
        if node is not None:
            print(node.data)
            self.preorder(node.left)
            self.preorder(node.right)

    def postorder(self, node):
        if node is not None:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data)


# Test
tree = Tree()

tree.insert(30)
tree.insert(10)
tree.insert(20)
tree.insert(40)
tree.insert(50)


tree.preorder(tree.getRoot())# Class for tree node
class Node:
    def __init__(self, data = None):
        self.data = data
        self.left = None
        self.right = None

# Class for tree
class Tree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def insert(self, data):
        new_node = Node(data)
        if self.root is None:
            self.root = new_node
        else:
            self.insertNode(self.root, new_node)

    def insertNode(self, node, new_node):
        if new_node.data < node.data:
            if node.left is None:
                node.left = new_node
            else:
                self.insertNode(node.left, new_node)
        else:
            if node.right is None:
                node.right = new_node
            else:
                self.insertNode(node.right, new_node)

    def inorder(self, node):
        if node is not None:
            self.inorder(node.left)
            print(node.data)
            self.inorder(node.right)

    def preorder(self, node):
        if node is not None:
            print(node.data)
            self.preorder(node.left)
            self.preorder(node.right)

    def postorder(self, node):
        if node is not None:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data)


# Test
tree = Tree()

tree.insert(30)
tree.insert(10)
tree.insert(20)
tree.insert(40)
tree.insert(50)

tree.inorder(tree.getRoot())

tree.preorder(tree.getRoot())

tree.postorder(tree.getRoot())
```
