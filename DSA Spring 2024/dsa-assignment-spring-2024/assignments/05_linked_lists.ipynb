In a Doubly linked list, each node contains a data element and two pointers, one pointing to the next node in the sequence and another pointing to the previous node.
It allows for traversal in both directions, forward and backward, to perform operations such as insertion and deletion at both ends of the list.

```python
# Class for linkedlist node
class Node:
    def __init__(self, data=None, next=None):
        self.prev = None
        self.data = data
        self.next = None


# Class for linkedlist
class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def add_node(self, data):
        # TODO: Add new Node to Double linked list.
        new_node = Node(data)
        if self.head:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current
            new_node.next = None
        else:
            self.head = new_node

        # self.display() # Last Line in method
        self.display()
        


    def insert_at_the_beginning(self, data):
        # TODO: Insert new Node at the beginning of Double linked list.
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            new_node.next.prev = new_node
            self.head = new_node

        # self.display() # Last Line in method
        self.display()

    def remove_node(self, target):
        # TODO: Remove existing Node from the Double linked list. Optional method
        if self.head is None:
            # print("Element " + str(target) + " does not exist in the list.") # Last Line in method
            print("Element " + str(target) + " does not exist in the list.")
            return
        
        current = self.head
        while current is not None:
            if current.data == target:
                print('on')
                if current.prev is None:
                    self.head = current.next
                    if current.next != None:
                        current.next.prev = None
                elif current.next is None:
                    if current.prev is not None:
                        current.prev.next = None
                else:
                    current.prev.next = current.next
                    current.next.prev = current.prev

                self.display()
                return
                

    def display(self):
        current = self.head
        nodes = []
        print('\n')
        while current:
            nodes.append(str(current.data))
            current = current.next
        print(' <-> '.join(nodes) + '\n')

    def to_string(self):
        current = self.head
        nodes = []
        while current:
            nodes.append(str(current.data))
            current = current.next
        return ' : '.join(nodes)


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def name(self):
        return self.name

    def age(self):
        return self.age

    def __str__(self):
        return f"{self.name}, {self.age}"


# Test
llist = DoubleLinkedList()

llist.add_node(Person("Bob", 45))
llist.add_node(Person("Jill", 37))
llist.add_node(Person("Jane", 29))
llist.insert_at_the_beginning(Person("Jack", 59))
# Optional part
# llist.remove_node(Person("Jack", 59))

llist.display()

result = "Jack, 59 : Bob, 45 : Jill, 37 : Jane, 29"

assert llist.to_string() == result

print(f"answer '{llist.to_string()}' was OK!")
```
